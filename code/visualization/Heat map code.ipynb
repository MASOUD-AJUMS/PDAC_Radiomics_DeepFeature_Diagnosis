```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Replace this file path with the location of your Excel file
file_path = r"e:\result step 1 label 2 radiomics with ROC and AUC\Label_2.xlsx_metrics.xlsx"  
# If your data is in a sheet with a different name, for example "Sheet1", you can use:
# df = pd.read_excel(file_path, sheet_name="Sheet1")
df = pd.read_excel(file_path)

# Ensure that your Excel file includes the columns:
# "Model", "Feature Selection", "Precision", "Recall", "F1-Score", "Accuracy", "AUC"

# Combine Model and Feature Selection into a new column for better y-axis labels
df["Model_FS"] = df["Model"] + "\n(" + df["Feature Selection"] + ")"

# Select the performance metric columns and convert them to percentages
metrics = ["Precision", "Recall", "F1-Score", "Accuracy", "AUC"]
heatmap_data = df.set_index("Model_FS")[metrics] * 100
heatmap_data = heatmap_data.round(2)

# Set a professional theme with custom fonts and bold styling for axes and labels
sns.set_theme(style="whitegrid", font_scale=1.2, rc={
    "axes.labelsize": 16,
    "axes.titlesize": 20,
    "axes.titleweight": "bold",
    "axes.labelweight": "bold",
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
    "font.family": "sans-serif",
    "font.weight": "bold"
})

plt.figure(figsize=(14, 20))

# Use a diverging palette for the heatmap
cmap = sns.diverging_palette(240, 10, n=9, as_cmap=True)

# Create the heatmap with annotations and enhanced styling
ax = sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap=cmap,
                 linewidths=0.5, linecolor='white',
                 cbar_kws={"shrink": 0.5, "label": "Percentage (%)"},
                 annot_kws={"fontsize": 12, "fontweight": "bold"})

# Add title and axis labels with enhanced styling
ax.set_title("Approach 1_Venis Label Radiomics Metrics Heatmap", pad=20)
ax.set_xlabel("Metrics", labelpad=10)
ax.set_ylabel("Model (Feature Selection)", labelpad=10)

# Customize tick labels for a bold and clean look
plt.xticks(rotation=45, ha='right', fontsize=14, fontweight="bold")
plt.yticks(rotation=0, fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
```
