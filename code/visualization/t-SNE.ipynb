```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import seaborn as sns

# Prompt the user for the Excel file path
file_path = r'e:\Label_5.xlsx'

# Load the Excel data
data = pd.read_excel(file_path)

# Ensure the Excel file contains a 'label' column
if 'label' not in data.columns:
    raise ValueError("The Excel file must contain a column named 'label'.")

# Separate labels from features
labels = data['label']
features = data.drop('label', axis=1)

# Apply t-SNE to reduce features to two dimensions
tsne = TSNE(n_components=2, random_state=992)
tsne_results = tsne.fit_transform(features)

# Create a DataFrame with the t-SNE results and labels
tsne_df = pd.DataFrame(tsne_results, columns=['TSNE1', 'TSNE2'])
tsne_df['label'] = labels.values

# Use the default style (white background) and set figure size
plt.style.use('default')
plt.figure(figsize=(8, 6))

# Get unique labels (assuming exactly two for blue and yellow)
unique_labels = tsne_df['label'].unique()
if len(unique_labels) != 2:
    print("Warning: Expected exactly two labels. Using the first two labels for coloring.")
    unique_labels = unique_labels[:2]

# Define the color mapping: first label as blue, second as yellow
color_dict = {unique_labels[1]:'yellow' , unique_labels[0]: 'blue'}

# Plot density contours (KDE) and overlay scatter points for each label group
for lbl in unique_labels:
    subset = tsne_df[tsne_df['label'] == lbl]
    
    # Draw KDE contours for density visualization
    sns.kdeplot(
        data=subset,
        x='TSNE1',
        y='TSNE2',
        fill=True,
        thresh=0,
        levels=10,
        alpha=0.5,
        color=color_dict[lbl],
        label=f'{lbl} density'
    )
    
    # Overlay the individual scatter points
    plt.scatter(
        subset['TSNE1'], subset['TSNE2'],
        s=50, alpha=0.7,
        color=color_dict[lbl],
        edgecolor='k', linewidth=0.5
    )

# Add grid lines with a light gray dashed style
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

# Customize titles, labels, and legend for a publication-ready look
#plt.title("t-SNE Visualization with Density Contours", fontsize=20)
#plt.xlabel("t-SNE Component 1", fontsize=14)
#plt.ylabel("t-SNE Component 2", fontsize=14)
#plt.legend(title="Label", fontsize=12, title_fontsize=14)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Replace these with the paths to your NIfTI files
ct_filename = r'c:\Users\alire\Downloads\100002_00001_0000.nii\100002_00001_0000.nii'
mask_filename = r'c:\Users\alire\Downloads\100002_00001.nii\100002_00001.nii'

# Load the CT image and mask
ct_img = nib.load(ct_filename)
mask_img = nib.load(mask_filename)

# Get the image data as numpy arrays
ct_data = ct_img.get_fdata()
mask_data = mask_img.get_fdata()

# ---- Visualization: Overlay on a mid-slice ----
# Choose a slice index (here we take the mid-slice along the axial direction)
slice_idx = ct_data.shape[2] // 2
ct_slice = ct_data[:, :, slice_idx]
mask_slice = mask_data[:, :, slice_idx]

# Plot the CT slice and overlay the mask with a semi-transparent colormap
plt.figure(figsize=(8, 8))
plt.imshow(ct_slice.T, cmap='gray', origin='lower')
# Mask overlay: display only where mask is nonzero
plt.imshow(np.ma.masked_where(mask_slice == 0, mask_slice).T, cmap='jet', alpha=0.5, origin='lower')
plt.title('CT Image with Mask Overlay')
plt.axis('off')
plt.show()

# ---- Create a new NIfTI overlay image ----
# Here we create an overlay volume by modifying the CT image where the mask is present.
# For example, we can highlight masked regions by setting them to the maximum intensity.
overlay_data = ct_data.copy()
max_intensity = np.max(ct_data)
overlay_data[mask_data > 0] = max_intensity

# Create a new NIfTI image using the same affine and header as the original CT
overlay_img = nib.Nifti1Image(overlay_data, ct_img.affine, ct_img.header)
overlay_filename = r'c:\Users\alire\Downloads\100002_00001_0000.nii\overlay_image.nii'
nib.save(overlay_img, overlay_filename)

print(f"Overlay image saved as {overlay_filename}")
```

```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Replace these with the paths to your NIfTI files
ct_filename = r'c:\Users\alire\Downloads\100002_00001_0000.nii\100002_00001_0000.nii'
mask_filename = r'c:\Users\alire\Downloads\100002_00001.nii\100002_00001.nii'

# Load the CT image and mask
ct_img = nib.load(ct_filename)
mask_img = nib.load(mask_filename)

# Get the image data as numpy arrays
ct_data = ct_img.get_fdata()
mask_data = mask_img.get_fdata()

# ---- Visualization: Display CT image, Mask, and Overlay ----
# Choose a slice index (here we take the mid-slice along the axial direction)
slice_idx = ct_data.shape[2] // 2
ct_slice = ct_data[:, :, slice_idx]
mask_slice = mask_data[:, :, slice_idx]

# Create a figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Display the CT image
axes[0].imshow(ct_slice.T, cmap='gray', origin='lower')
axes[0].set_title('CT Image')
axes[0].axis('on') # Show axis ticks if you want to align the grid
axes[0].grid(True, color='white', linestyle='--', linewidth=0.5)

axes[1].imshow(mask_slice.T, cmap='viridis', origin='lower')
axes[1].set_title('Mask')
axes[1].axis('on')
axes[1].grid(True, color='white', linestyle='--', linewidth=0.5)

axes[2].imshow(ct_slice.T, cmap='gray', origin='lower')
axes[2].imshow(np.ma.masked_where(mask_slice == 0, mask_slice).T,
cmap='jet', alpha=0.5, origin='lower')
axes[2].set_title('CT Image with Mask Overlay')
axes[2].axis('on')
axes[2].grid(True, color='white', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()

# ---- Create a new NIfTI overlay image ----
# Here we create an overlay volume by modifying the CT image where the mask is present.
# For example, we can highlight masked regions by setting them to the maximum intensity.
overlay_data = ct_data.copy()
max_intensity = np.max(ct_data)
overlay_data[mask_data > 0] = max_intensity

# Create a new NIfTI image using the same affine and header as the original CT
overlay_img = nib.Nifti1Image(overlay_data, ct_img.affine, ct_img.header)
overlay_filename = r'c:\Users\alire\Downloads\100002_00001_0000.nii\overlay_image.nii'
nib.save(overlay_img, overlay_filename)

print(f"Overlay image saved as {overlay_filename}")
```
