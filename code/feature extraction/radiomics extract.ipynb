# کد برای پروژه پانکراس

```
import os
import numpy as np
import pandas as pd
import SimpleITK as sitk
from radiomics import featureextractor
import radiomics

# Set the paths to the directory containing images and masks
image_path = '/content/100002_00001_0000.nii.gz'
mask_path = '/content/100002_00001.nii.gz'

# Create empty lists to store the image names and features
image_names = []
features = []

# Read the image and mask using SimpleITK
image = sitk.ReadImage(image_path)
mask = sitk.ReadImage(mask_path)

# Convert the mask to a supported type (unsigned 32-bit integer)
mask = sitk.Cast(mask, sitk.sitkUInt32)

# Convert the mask to a numpy array to get the unique labels
mask_array = sitk.GetArrayFromImage(mask)
unique_labels = np.unique(mask_array)

# Initialize the feature extractor (without specifying the label here)
settings = {'additionalInfo': True}
extractor = radiomics.featureextractor.RadiomicsFeatureExtractor(**settings)

# Loop over all unique labels and extract features for each one
for label in unique_labels:
    if label == 0:  # Skip the background label
        continue

    # Check if the label exists in the mask
    if np.any(mask_array == label):
        print(f"Extracting features for Label {label}")
        result = extractor.execute(image, mask, label=int(label))

        # Append the image name and features to the lists
        image_names.append(f'Label_{label}')
        features.append(result)

# Convert the result dictionary to a DataFrame
df = pd.DataFrame(features, index=image_names)

# Set the output path for the CSV file
output_path = '/content/drive/MyDrive/pyradiomics/ai_test_all_labels2.csv'

# Save the DataFrame to a CSV file
df.to_csv(output_path)
```

## **کد برای پروژه  پانکراس _2**

```
import numpy as np
import pandas as pd
import SimpleITK as sitk
from radiomics import featureextractor

# Load the image and mask
image = sitk.ReadImage(image_path)
mask = sitk.ReadImage(mask_path)

# تبدیل ماسک به آرایه numpy برای پیدا کردن لیبل‌ها
mask_array = sitk.GetArrayFromImage(mask)
available_labels = np.unique(mask_array)

# لیست اسامی لیبل‌ها (در اینجا شامل 1 تا 6)
individual_labels = [1, 2, 3, 4, 5, 6]

# ایجاد لیست خالی برای ذخیره ویژگی‌ها و برچسب‌ها
features = []
labels = []

# تنظیمات PyRadiomics
settings = {
    'additionalInfo': True,
    'enabledImageTypes': {'Original': {}, 'Wavelet': {}},  # فعال‌سازی موجک‌ها و تصویر اصلی
    'enabledFeatures': {
        'firstorder': [],  # First order statistics
        'glcm': [],        # GLCM features
        'gldm': [],        # GLDM features
        'glrlm': [],       # GLRLM features
        'glszm': [],       # GLSZM features
        'ngtdm': [],       # NGTDM features,
        'shape': []        # Shape-based features
    }
}
extractor = featureextractor.RadiomicsFeatureExtractor(**settings)

# استخراج ویژگی‌ها برای هر لیبل
for label in individual_labels:
    if label in available_labels:  # بررسی اینکه آیا لیبل در ماسک موجود است
        print(f"Extracting features for Label {label}")
        result = extractor.execute(image, mask, label=int(label))

        # ذخیره ویژگی‌ها
        features.append(result)

        # برچسب (مثال: افراد سالم برچسب 0 و بیماران برچسب 1)
        if label in [1, 5]:  # اگر لیبل متعلق به بیماران باشد
            labels.append(1)
        else:  # لیبل‌های افراد سالم
            labels.append(0)
    else:
        print(f"Label {label} not present in the mask.")

# تبدیل ویژگی‌ها به DataFrame
df_features = pd.DataFrame(features)

# افزودن برچسب‌ها به DataFrame
df_features['label'] = labels

# نمایش DataFrame نهایی
print(df_features)
output_path = '/content/drive/MyDrive/pyradiomics/ai_testcc.csv'

# Save the DataFrame to a CSV file
df_features.to_csv(output_path)
```

```
import os
import pandas as pd
import SimpleITK as sitk
from radiomics import featureextractor

# Set the paths to the directory containing images and masks
image_dir = '/content/drive/MyDrive/pyradiomics/images'
mask_dir = '/content/drive/MyDrive/pyradiomics/mask'

# Initialize the feature extractor with additional parameters
params = {
    'binWidth': 20,
    'sigma': [1, 2, 3],
    'verbose': True,
    'normalize': True,
    'ignoreZeroVoxel': True,
    'additionalInfo': False,
    'resampledPixelSpacing': None,
    'label': 2,
}

# Specify the filters to be applied
image_types = {
    "Original": {},
    "Wavelet": {},
    "LoG": {"sigma": [1, 2, 3]},
    "Square": {},
    "SquareRoot": {},
    "Logarithm": {},
    "Exponential": {},
    "Gradient": {},
    "LBP2D": {},
    "LBP3D": {},
    "SquareRoot": {},
    "Logarithm": {},
    "Exponential": {}
}

# Initialize the feature extractor
extractor = featureextractor.RadiomicsFeatureExtractor(**params)
extractor.enableAllImageTypes()  # Enable all default image types

# Enable specific filters
for image_type, settings in image_types.items():
    extractor.enableImageTypeByName(image_type, settings)

# Create empty lists to store the image names and features
image_names = []
features = []

# Iterate over the images and masks in the directories
for image_file in os.listdir(image_dir):
    if image_file.endswith('.nii.gz'):
        image_path = os.path.join(image_dir, image_file)
        mask_file = image_file.replace('image', 'mask')
        mask_path = os.path.join(mask_dir, mask_file)

        if os.path.exists(mask_path):
            # Read the image and mask
            image = sitk.ReadImage(image_path)
            mask = sitk.ReadImage(mask_path)

            # Extract the features
            result = extractor.execute(image, mask)

            # Append the image name and features to the lists
            image_names.append(image_file)
            features.append(result)

# Create a DataFrame from the features
df = pd.DataFrame(features, index=image_names)

# Save the features as a CSV file
output_path2 = '/content/drive/MyDrive/pyradiomics/test288.csv'
df.to_csv(output_path2)
```
